```python
from sklearn.datasets import make_circles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X1, y1 = make_circles(n_samples=30_000, random_state=0, noise=0.04)
X2, y2 = make_circles(n_samples=30_000, random_state=1, noise=0.04)
X2[:, 0] += 0.6
X_info = np.concatenate((X1, X2))
X = X_info
y = np.concatenate((y1, y2 + 2))

w = 2
fig, ax = plt.subplots(figsize=(w, 0.8 * w))
for c in np.unique(y):
    i = y == c
    X_show = X[i]
    X_show = X_show[:4000]
    ax.scatter(X_show[:, 0], X_show[:, 1], alpha=0.007)
# print(1.8 + 1.3)
# print(2 * 1.55)
# ax.set_xlim(-1.3, 1.8)
# ax.set_ylim(-1.55, 1.55)
    
    
plt.savefig("figs/synthetic-dataset.png", bbox_inches="tight")
```
