From Figure 2 of http://proceedings.mlr.press/v80/falkner18a/falkner18a.pdf

```python
data = [
    {"num_workers": 1, "time_to_regret": 2000},
    {"num_workers": 2, "time_to_regret": 1000},
    {"num_workers": 4, "time_to_regret": 575},
    {"num_workers": 8, "time_to_regret": 300},
    {"num_workers": 16, "time_to_regret": 200},
    {"num_workers": 32, "time_to_regret": 130},
]
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(data)
df["speedup"] = df.time_to_regret.max() / df["time_to_regret"]

ax = df.plot(x="num_workers", y="speedup")
ax.set_yscale("log", basey=2)
ax.set_xscale("log", basex=2)
ax.grid()
x = np.logspace(0, np.log10(32))
ax.plot(x, x)
```
