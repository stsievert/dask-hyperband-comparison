```python
from benchmark import dataset
(X, y), _ = dataset()
```

```python
import pandas as pd
df = pd.DataFrame(X)
df["y"] = y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

w = 3.0
fig, ax = plt.subplots(figsize=(w, w))
ax = sns.scatterplot(
    x=1,
    y=0,
    hue="y",
    data=df.sample(20_000),
    linewidth=0,
    alpha=0.1,
    palette="Set1",
)
ax.legend_.remove()
ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
```
